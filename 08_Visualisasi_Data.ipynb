```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Menghasilkan data acak untuk berbagai contoh
np.random.seed(42)
data = pd.DataFrame({
    'time': pd.date_range(start='2025-01-01', periods=100, freq='D'),
    'category': np.random.choice(['A', 'B', 'C', 'D'], size=100),
    'value1': np.random.rand(100) * 100,
    'value2': np.random.rand(100) * 50 + 50,
    'x': np.random.randn(100),
    'y': np.random.randn(100)
})
sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(10,4))
sns.lineplot(x='time', y='value1', data=data, marker='o')
plt.title('Line Chart - Tren Nilai Harian')
plt.xlabel('Tanggal')
plt.ylabel('Nilai Sensor')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x='category', y='value1', data=data, estimator=np.mean, palette='viridis')
plt.title('Bar Chart - Rata-rata Nilai per Kategori')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.barplot(y='category', x='value1', data=data, estimator=np.mean, palette='coolwarm')
plt.title('Column Chart - Rata-rata Nilai per Kategori (Horizontal)')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x='value1', y='value2', hue='category', data=data, s=70)
plt.title('Scatter Plot - Hubungan value1 dan value2')
plt.show()
```

```python
plt.figure(figsize=(6,5))
corr = data[['value1', 'value2', 'x', 'y']].corr()
sns.heatmap(corr, annot=True, cmap='YlGnBu')
plt.title('Heatmap Korelasi antar Variabel')
plt.show()
```

```python
import pandas as pd
import numpy as np
import plotly.express as px

# Contoh data lokasi (simulasi koordinat sensor IoT di Lombok)
np.random.seed(42)
geo = pd.DataFrame({
    'latitude': np.random.uniform(-8.7, -8.4, 50),
    'longitude': np.random.uniform(116.0, 116.3, 50),
    'sensor_type': np.random.choice(['Suhu', 'Kelembaban', 'Tekanan'], 50),
    'value': np.random.randint(20, 100, 50)
})

# Membuat Geo Map
fig = px.scatter_mapbox(
    geo,
    lat='latitude',
    lon='longitude',
    color='sensor_type',
    size='value',
    hover_name='sensor_type',
    zoom=9,
    mapbox_style='carto-positron',
    title='Peta Sensor IoT Wilayah Lombok (Simulasi)'
)
fig.show()
```

```python
# Simulasi adjacency matrix (hubungan antar node)
nodes = ['A','B','C','D']
relations = pd.DataFrame(np.random.randint(0,10,(4,4)), columns=nodes, index=nodes)
plt.figure(figsize=(5,4))
sns.heatmap(relations, annot=True, cmap='coolwarm')
plt.title('Network Relationship Matrix (Simulasi Hubungan Node)')
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Simulasi jaringan IoT antar node sensor
G = nx.Graph()

# Node (misal: gateway dan sensor)
nodes = ['Gateway', 'Sensor1', 'Sensor2', 'Sensor3', 'Sensor4']
G.add_nodes_from(nodes)

# Edge (hubungan komunikasi)
edges = [
    ('Gateway', 'Sensor1'),
    ('Gateway', 'Sensor2'),
    ('Sensor1', 'Sensor3'),
    ('Sensor2', 'Sensor4'),
    ('Sensor3', 'Sensor4')
]
G.add_edges_from(edges)

# Visualisasi graf
plt.figure(figsize=(6,5))
pos = nx.spring_layout(G, seed=42)
nx.draw(G, pos, with_labels=True, node_color='skyblue', node_size=2000, edge_color='gray', font_size=10, font_weight='bold')
plt.title('Network Graph - Hubungan Antar Node IoT')
plt.show()
```

```python
hierarki = pd.DataFrame({
    'Kelompok': np.random.choice(['Sensor', 'Aktuator'], 20),
    'SubKategori': np.random.choice(['Suhu', 'Kelembaban', 'Tekanan'], 20),
    'Nilai': np.random.randint(10, 100, 20)
})
plt.figure(figsize=(8,4))
sns.barplot(x='SubKategori', y='Nilai', hue='Kelompok', data=hierarki)
plt.title('Simulasi Tree/Hierarchy Visualization dengan Bar Plot')
plt.show()
```

```python
import plotly.express as px
import pandas as pd

# Contoh data hierarki
tree = pd.DataFrame({
    'Kelompok': ['Sensor']*6 + ['Aktuator']*4,
    'SubKategori': ['Suhu', 'Kelembaban', 'Cahaya', 'Tekanan', 'Air', 'Tanah', 'Motor', 'Relay', 'Kipas', 'Valve'],
    'Nilai': [120, 95, 80, 75, 60, 50, 100, 85, 65, 70]
})

# Tree map
fig = px.treemap(
    tree,
    path=['Kelompok', 'SubKategori'],
    values='Nilai',
    color='Nilai',
    color_continuous_scale='viridis',
    title='Tree Map - Hierarki Sensor dan Aktuator IoT'
)
fig.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='category', y='value1', data=data, palette='Set2')
plt.title('Box Plot - Distribusi Data per Kategori')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.violinplot(x='category', y='value2', data=data, inner='quartile', palette='muted')
plt.title('Violin Plot - Distribusi value2 per Kategori')
plt.show()
```

```python
from IPython.display import clear_output
import time

values = []
plt.figure(figsize=(8,4))

for i in range(30):
    new_val = np.random.randn() + 10
    values.append(new_val)
    clear_output(wait=True)
    sns.lineplot(x=range(len(values)), y=values)
    plt.title(f'Streaming Data (Iterasi {i+1})')
    plt.xlabel('Waktu')
    plt.ylabel('Nilai Sensor')
    plt.show()
    time.sleep(0.2)
```
